```python
Flask HTTP Methods

HTTP stands for Hyper Text Transfer Protocol which is considered for the fundamental concepts of data transurLfer in the WWW(World Wide Web).
All Web technologies including Flask need to provide several HTTP Methods for data communication.

____________________________________________________________________________________________________________________________________
|SNo|  Method   |     Description                                                                                                  |
|___|___________|__________________________________________________________________________________________________________________|
|1  | GET       | It is the most common method which can be used to send data in unencrypted form to the server.                   |
|___|___________|__________________________________________________________________________________________________________________|
|2  | HEAD      | It is Similar to the Get but without the response body.                                                          |
|___|___________|__________________________________________________________________________________________________________________|
|3  | POST      | It is used to send the form data to the server. The server does not cache the data transmitted using post method.|
|___|___________|__________________________________________________________________________________________________________________|
|4  | PUT       | It is used to replace all the current representation of the target resource with the upload content.             |
|___|___________|__________________________________________________________________________________________________________________|
|5  | DELETE    | It is used to delete all the current representation of the target resorce with specified URL.                    |
|___|___________|__________________________________________________________________________________________________________________|
```

```python
from flask import Flask, request, render_template_string
app = Flask(__name__)
@app.route('/', methods=['GET'])
def index():
    form_html = """
    <html>
        <body>
            <form action="/login" method="POST">
                <table>
                    <tr>
                        <td>Name</td>
                        <td><input type="text" name="username"></td>
                    </tr>
                    <tr>
                        <td>Password</td>
                        <td><input type="password" name="password"></td>
                    </tr>
                    <tr>
                        <td><input type="submit"></td>
                    </tr>
                </table>
            </form>
        </body>
    </html>
    """
    return render_template_string(form_html)
@app.route('/login', methods=['POST'])
def login():
    username = request.form.get('username')  
    password = request.form.get('password')
    if username == 'balaji' and password == 'balaji9950':
        return f"Welcome to my Flask website, {username}"
if __name__ == '__main__':
    app.run(port=5200)
```

```python
Delimiters

Jinja2 template engine provides some delimiters which can be used in the html to make it capable of dynamic data representation.
This Template system provides some HTML syntax which are placeholders for variables and expressions that are replaced by the actual
values when the template is rendered.

The jinja2 template engine provides the following delimiters to escape from the HTML.

1) {%....%} - for statements
2) {{...}}  - for expressions to print the template output.
3) {#...#}  - for the comments that are not included in the template output.
4) #....#   - for line statements
```
